## dfのみで処理

```python
def pd_to_expected(word,num,df):
    comma='.'
    name=comma*num+word+comma*(4-num)
    total=len(df)
    n_n=len(df[df['名前'].str.contains(word)])
    n_g=len(df.query('名前.str.contains("'+name+'")',engine='python'))
    expected=((total-n_n)/total*n_n+
              (n_n-n_g)/total*(total-n_n+n_g)+
              n_g/total*(total-n_g)
             )
    return expected
```

```python
def poke_to_expected(poke,df):
    value=0
    char=""
    if len(poke)==len(set(poke)):
        for i in range(0,5):
            value+=pd_to_expected(poke[i],i,df)
    else:
        for i in range(0,5):
            value+=pd_to_expected(poke[i],i,df)
            if list(poke).count(i)==2:
                char=i
        value-=len(df[df['名前'].str.contains(char)])
    return value
```

```python
def make_df_expected(df):
    dict_expected={}
    for i in range(0,len(df)):
        dict_expected[df.iloc[i,1]]=[poke_to_expected(df.iloc[i,1],df)]
    df_expected=pd.DataFrame(dict_expected)
    df_expected=df_expected.T.rename(columns={0:'期待値'}).sort_values('期待値',ascending=False)
    return df_expected
```

```python
def cut_df(poke,color,df):
    comma='.'
    double=''
    double2=''
    if len(list(poke))==len(set(poke)):
        for i in range(0,5):
            if color[i]=='n':
                cut_df=df[df['名前'].str.contains(poke[i])]
                and_list=list(set(df.index) & set(cut_df.index))
                df=df.drop(and_list)
            elif color[i]=='y':
                name=comma*i+poke[i]+comma*(4-i)
                cut_df=df.query('名前.str.contains("'+name+'")',engine='python')
                df=df[df['名前'].str.contains(poke[i])]
                and_list=list(set(df.index) & set(cut_df.index))
                df=df.drop(and_list)
            elif color[i]=='g':
                name=comma*i+poke[i]+comma*(4-i)
                df=df.query('名前.str.contains("'+name+'")',engine='python')
    elif len(list(poke))-len(set(poke))>=1:
        for i in range(0,5):
            if list(poke).count(poke[i])==1:
                if color[i]=='n':
                    cut_df=df[df['名前'].str.contains(poke[i])]
                    and_list=list(set(df.index) & set(cut_df.index))
                    df=df.drop(and_list)
                elif color[i]=='y':
                    name=comma*i+poke[i]+comma*(4-i)
                    cut_df=df.query('名前.str.contains("'+name+'")',engine='python')
                    df=df[df['名前'].str.contains(poke[i])]
                    and_list=list(set(df.index) & set(cut_df.index))
                    df=df.drop(and_list)
                elif color[i]=='g':
                    name=comma*i+poke[i]+comma*(4-i)
                    df=df.query('名前.str.contains("'+name+'")',engine='python')
            elif list(poke).count(poke[i])==2:
                if double=='':
                    double=poke[i]
                    list_index=[i for i, d in enumerate(list(poke)) if d == double]
                    
                    if color[list_index[0]]+color[list_index[1]] == 'yn':
                        name=comma*list_index[0]+poke[list_index[0]]+comma*(4-list_index[0])
                        cut_df=df.query('名前.str.contains("'+name+'")',engine='python')
                        df=df[df['名前'].str.contains(poke[list_index[0]])]
                        and_list=list(set(df.index) & set(cut_df.index))
                        df=df.drop(and_list)
                    elif color[list_index[0]]+color[list_index[1]] == 'gn':
                        name=comma*list_index[0]+poke[list_index[0]]+comma*(4-list_index[0])
                        df=df.query('名前.str.contains("'+name+'")',engine='python')
                    elif color[list_index[0]]+color[list_index[1]] == 'ng':
                        name=comma*list_index[1]+poke[list_index[1]]+comma*(4-list_index[1])
                        df=df.query('名前.str.contains("'+name+'")',engine='python')
                elif double!=poke[i] and double2=='':
                    double2=poke[i]
                    list_index=[i for i, d in enumerate(list(poke)) if d == double2]
                    if color[list_index[0]]+color[list_index[1]] == 'yn':
                        name=comma*list_index[0]+poke[list_index[0]]+comma*(4-list_index[0])
                        cut_df=df.query('名前.str.contains("'+name+'")',engine='python')
                        df=df[df['名前'].str.contains(poke[list_index[0]])]
                        and_list=list(set(df.index) & set(cut_df.index))
                        df=df.drop(and_list)
                    elif color[list_index[0]]+color[list_index[1]] == 'gn':
                        name=comma*list_index[0]+poke[list_index[0]]+comma*(4-list_index[0])
                        df=df.query('名前.str.contains("'+name+'")',engine='python')
                    elif color[list_index[0]]+color[list_index[1]] == 'ng':
                        name=comma*list_index[1]+poke[list_index[1]]+comma*(4-list_index[1])
                        df=df.query('名前.str.contains("'+name+'")',engine='python')
            elif list(poke).count(poke[i])==3:
                if double=='':
                    double=poke[i]
                    list_index=[i for i, d in enumerate(list(poke)) if d == double]
                    
                    if color[list_index[0]]+color[list_index[1]] == 'yn':
                        name=comma*list_index[0]+poke[list_index[0]]+comma*(4-list_index[0])
                        cut_df=df.query('名前.str.contains("'+name+'")',engine='python')
                        df=df[df['名前'].str.contains(poke[list_index[0]])]
                        and_list=list(set(df.index) & set(cut_df.index))
                        df=df.drop(and_list)
                    elif color[list_index[0]]+color[list_index[1]] == 'gn':
                        name=comma*list_index[0]+poke[list_index[0]]+comma*(4-list_index[0])
                        df=df.query('名前.str.contains("'+name+'")',engine='python')
                    elif color[list_index[0]]+color[list_index[1]] == 'ng':
                        name=comma*list_index[1]+poke[list_index[1]]+comma*(4-list_index[1])
                        df=df.query('名前.str.contains("'+name+'")',engine='python')                
    return df
```

```python
def solve_wordle(df):
    global list_cut
    list_cut=[]
    t1=time.time()
    df_cut=df.copy()
    total=len(df)
    n=0
    n_challenge=0
    reset_first=0
    poke=make_df_expected(df).index[0]
    poke_shote=make_df_expected(df).index[0]
    print(poke+'を記入します。')
    while len(df)>0:
        n_challenge+=1
        write_poke(poke)
        button_enter.click()
        time.sleep(4)
        color=''
        for i in range(100+n*5,105+n*5):
            if 'green' in items[i].find_elements_by_xpath("./div")[0].get_attribute("class"):
                color += 'g'
            elif 'grey' in items[i].find_elements_by_xpath("./div")[0].get_attribute("class"):
                color += 'n'
            else:
                color += 'y'
        print('色は'+color+'です。')
        if color=='ggggg':
            n=0
            list_cut.append(poke)
            df=df.drop([df[df['名前']==poke].index[0]])
            sucess=total-len(df)
            try:
                button_ok.click()
            except:
                print('クリアしました！')
                break
            print(str(sucess)+'匹正解しました。')
            if sucess < 150 and sucess % 5 ==0:
                poke_shote=make_df_expected(df).index[0]
                poke=poke_shote
            elif sucess >=150:
                poke=make_df_expected(df).index[0]
            else:
                poke=poke_shote
                
            print(poke+'を記入します。')
            df_cut=df.copy()
        else:
            n+=1
            if df_cut.equals(df):
                df_cut=cut_df(poke,color,df)
                poke=make_df_expected(df_cut).index[0]
                print(poke+'を記入します。')
            else:
                df_cut=cut_df(poke,color,df_cut)
                poke=make_df_expected(df_cut).index[0]
                print(poke+'を記入します。')
    t2=time.time()
    print(str(n_challenge)+'回入力しました。')
    total_time=t2-t1
    print(str(total_time)+'秒かかりました。')
```

```python
#初期値はダイパまででサシカマス
solve_wordle(df)
```
